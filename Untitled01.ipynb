```python
# prompt: can you find out basic stats information such as mean, median , mode, variance , std , skewness, kurtosis , range

import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

df = pd.read_csv('Salaries.csv')

# Calculate the mean
mean_salary = df['salary'].mean()

# Calculate the median
median_salary = df['salary'].median()

# Calculate the mode
mode_salary = df['salary'].mode()[0]  # Access the first element if multiple modes exist

# Calculate the variance
variance_salary = df['salary'].var()

# Calculate the standard deviation
std_salary = df['salary'].std()

# Calculate the skewness
skewness_salary = skew(df['salary'])

# Calculate the kurtosis
kurtosis_salary = kurtosis(df['salary'])

# Calculate the range
range_salary = df['salary'].max() - df['salary'].min()


print(f"Mean: {mean_salary}")
print(f"Median: {median_salary}")
print(f"Mode: {mode_salary}")
print(f"Variance: {variance_salary}")
print(f"Standard Deviation: {std_salary}")
print(f"Skewness: {skewness_salary}")
print(f"Kurtosis: {kurtosis_salary}")
print(f"Range: {range_salary}")
```
